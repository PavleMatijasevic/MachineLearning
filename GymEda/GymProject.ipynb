#Inzenjering podataka

```python
# Dodavanje novih karakteristika

# Kreiranje kategorije na osnovu BMI
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

dataset["BMI_Category"] = dataset["BMI"].apply(bmi_category)

#  Dodavanje odnosa tezine i visine
dataset["Weight_Height_Ratio"] = dataset["Weight (kg)"] / dataset["Height (m)"]

# 3. Kategorisanje trajanja sesije na osnovu intervala
def session_duration_category(hours):
    if hours < 1:
        return "Short"
    elif 1 <= hours <= 2:
        return "Moderate"
    else:
        return "Long"

dataset["Session_Duration_Category"] = dataset["Session_Duration (hours)"].apply(session_duration_category)

# Prikaz azuriranog skupa
display(dataset.head())
```

Analiza distribucije novih karakteristika i potencijalnih outlier-a:

#Obuka regresionog modela

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler

target = "Calories_Burned"
features = dataset.drop(columns=[target])

# Kodiranje kategorijalnih varijabli
categorical_columns = features.select_dtypes(include=["object"]).columns
encoder = OneHotEncoder(sparse_output=False, drop="first")
encoded_categorical = encoder.fit_transform(features[categorical_columns])

# Dodavanje kodiranih kategorijalnih kolona
encoded_feature_names = encoder.get_feature_names_out(categorical_columns)
encoded_categorical_df = pd.DataFrame(encoded_categorical, columns=encoded_feature_names)
features = features.drop(columns=categorical_columns).reset_index(drop=True)
features = pd.concat([features, encoded_categorical_df], axis=1)

# Skaliranje numerickih podataka
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_features_df = pd.DataFrame(scaled_features, columns=features.columns)


X_train, X_test, y_train, y_test = train_test_split(
    scaled_features_df, dataset[target], test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

#evaluacija modela
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
```

Vizuelizacija performansi modela


```python
# Scatterplot za stvarne i predvidjene vrednosti
plt.figure(figsize=(8, 5))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.7)
plt.title("Actual vs Predicted Calories Burned")
plt.xlabel("Actual Calories Burned")
plt.ylabel("Predicted Calories Burned")
plt.axline([0, 0], [1, 1], color="red", linestyle="--", label="Perfect Prediction")
plt.legend()
plt.show()
```

Dodatni regresioni algoritmi

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor

# Lista modela
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Lasso Regression": Lasso(alpha=0.1),
    "Decision Tree Regressor": DecisionTreeRegressor(max_depth=5, random_state=42)
}


results = {}

for name, model in models.items():
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {"MSE": mse, "R2": r2}
    print(f"{name}: MSE = {mse:.2f}, R2 = {r2:.2f}")
```

Vizuelizacija rezultata

```python
# Vizualizacija MSE i R2 rezultata
results_df = pd.DataFrame(results).T

plt.figure(figsize=(8, 5))
sns.barplot(x=results_df.index, y="MSE", data=results_df.reset_index())
plt.title("Mean Squared Error (MSE) Comparison")
plt.ylabel("MSE")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x=results_df.index, y="R2", data=results_df.reset_index())
plt.title("R-Squared (R2) Comparison")
plt.ylabel("R2")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.show()
```

Izracunavanje znacajnosti obelezja

```python
# Znacajnost obelezja za Linear Regression model
feature_names = X_train.columns
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

coefficients = pd.Series(linear_model.coef_, index=feature_names)

# Grupisanje znacajnosti za kategorijalne karakteristike (OneHotEncoder komponente)
categorical_features = encoder.get_feature_names_out(categorical_columns)
categorical_importance = coefficients[categorical_features].groupby(lambda x: x.split('_')[0]).sum()

# Kombinovanje numerickih i kategorijalnih znacajnosti
total_importance = pd.concat([coefficients[~coefficients.index.isin(categorical_features)], categorical_importance])

# Sortiranje po znacajnosti
total_importance = total_importance.sort_values(ascending=False)

# Prikaz znacajnosti
print("Znacajnost obelezja (Linear Regression):")
display(total_importance)

# Vizualizacija znacajnosti
plt.figure(figsize=(10, 6))
total_importance.plot(kind='bar')
plt.title("Feature Importance (Linear Regression)")
plt.ylabel("Importance")
plt.xlabel("Features")
plt.show()
```

```python
# za nelinearni model

from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

feature_importance_rf = pd.Series(rf_model.feature_importances_, index=feature_names).sort_values(ascending=False)

print("Znacajnost obelezja (Random Forest):")
display(feature_importance_rf)


plt.figure(figsize=(10, 6))
feature_importance_rf.plot(kind='bar', color="skyblue")
plt.title("Feature Importance (Random Forest)")
plt.ylabel("Importance")
plt.xlabel("Features")
plt.show()
```

#Rad sa outlierima

```python

# Ponovno racunanje IQR za Weight-to-Height Ratio
q1 = dataset["Weight_Height_Ratio"].quantile(0.25)
q3 = dataset["Weight_Height_Ratio"].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Filtiranje outliera
outliers = dataset[(dataset["Weight_Height_Ratio"] < lower_bound) | (dataset["Weight_Height_Ratio"] > upper_bound)]
print(f"Broj outlier-a u Weight-to-Height Ratio: {len(outliers)}")
display(outliers)
```

```python
#Vizuelizacija outliera

# Scatterplot Weight-to-Height Ratio vs Calories_Burned (sa outlierima)
plt.figure(figsize=(8, 5))
sns.scatterplot(data=dataset, x="Weight_Height_Ratio", y="Calories_Burned", hue=dataset["Weight_Height_Ratio"] > upper_bound)
plt.title("Anomalies in Weight-to-Height Ratio vs Calories Burned")
plt.xlabel("Weight-to-Height Ratio")
plt.ylabel("Calories Burned")
plt.show()

# Boxplot za Weight-to-Height Ratio
plt.figure(figsize=(8, 5))
sns.boxplot(x=dataset["Weight_Height_Ratio"], color="orange")
plt.title("Boxplot: Weight-to-Height Ratio (Anomalies Highlighted)")
plt.show()
```

```python
#Uklanjanje outliera

cleaned_dataset = dataset[
    (dataset["Weight_Height_Ratio"] >= lower_bound) &
    (dataset["Weight_Height_Ratio"] <= upper_bound)
]
print(f"Broj redova nakon uklanjanja outliera: {len(cleaned_dataset)}")
```

Treniranje modela sa i bez anomalija

```python
# Identifikacija kategorijalnih kolona
categorical_columns = ["Gender", "Workout_Type"]
numeric_columns = cleaned_dataset.select_dtypes(include=[np.number]).columns

# One-Hot Encoding za kategorijalne kolone
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop="first")
encoded_categories = encoder.fit_transform(cleaned_dataset[categorical_columns])

# Pretvaranje u DataFrame
encoded_categories_df = pd.DataFrame(
    encoded_categories,
    columns=encoder.get_feature_names_out(categorical_columns)
)

# Spajanje numerikih i enkodiranih podataka
cleaned_dataset_numeric = pd.concat(
    [cleaned_dataset[numeric_columns].reset_index(drop=True),
     encoded_categories_df.reset_index(drop=True)],
    axis=1
)

# Ciljna varijabla (y) ostaje nepromenjena
X_cleaned = cleaned_dataset_numeric.drop(columns=["Calories_Burned"])
y_cleaned = cleaned_dataset["Calories_Burned"]
```

```python
# Podela na trening i test skupove
from sklearn.model_selection import train_test_split
X_train_cleaned, X_test_cleaned, y_train_cleaned, y_test_cleaned = train_test_split(
    X_cleaned, y_cleaned, test_size=0.2, random_state=42
)
```

```python
# Treniranje modela na dataset-u bez outliera
model_without_anomalies = LinearRegression()
model_without_anomalies.fit(X_train_cleaned, y_train_cleaned)
y_pred_without_anomalies = model_without_anomalies.predict(X_test_cleaned)


mse_without_anomalies = mean_squared_error(y_test_cleaned, y_pred_without_anomalies)
r2_without_anomalies = r2_score(y_test_cleaned, y_pred_without_anomalies)

print(f"MSE bez anomalija: {mse_without_anomalies:.2f}, R2 bez anomalija: {r2_without_anomalies:.2f}")
```
